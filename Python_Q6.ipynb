```python
import requests
import pandas as pd
import numpy as np
```

```python
link='https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
responce=requests.get(link)
data=responce.json()
data
```

```python
pokemon_list=data['pokemon']
```

```python
#Q1 : Get all pokemons whose spawn rate is less than 50%
spawn_rate_filtered = [pokemon for pokemon in pokemon_list if float(pokemon['spawn_chance']) < 5]
for pokemon in spawn_rate_filtered:
    print(pokemon['name'])
```

```python
# Bar plot for question 1: Pokemons with spawn rate less than 5%
import matplotlib.pyplot as plt
pokemon_names= [pokemon['name'] for pokemon in spawn_rate_filtered]
spawn_chances = [float(pokemon['spawn_chance']) for pokemon in spawn_rate_filtered]

fig=plt.figure(figsize =(12,6))
plt.bar(pokemon_names , spawn_chances)
plt.xlabel('pokemon')
plt.ylabel('spawn chance (%)')
plt.title('pokemons with spawn rate less than 5%')
plt.xticks(rotation=90, fontsize=8)
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()
plt.show()
```

```python
#Q2 : Get all Pokemons that have less than 4 weaknesses
weakness_filtered=[pokemon for pokemon in pokemon_list if len(pokemon['weaknesses']) < 4]
for pokemon in weakness_filtered:
    print(pokemon['name'])
```

```python
# Bar plot for question 2: Pokemons with less than 4 weaknesses
pokemon_names = [pokemon["name"] for pokemon in weakness_filtered]
weakness_counts = [len(pokemon["weaknesses"]) for pokemon in weakness_filtered]

fig=plt.figure(figsize =(12,6))
plt.bar(pokemon_names, weakness_counts)
plt.xlabel("Pokemon")
plt.ylabel("Number of Weaknesses")
plt.title("Pokemons with Less than 4 Weaknesses")
plt.xticks(rotation=90, fontsize=8)
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()
plt.show()
```

```python
#Q3 : Get all Pokemons that have no multipliers at all
no_multipliers=[pokemon for pokemon in pokemon_list if "multipliers" not in pokemon]
for pokemon in no_multipliers:
    print(pokemon['name'])
```

```python
# Pie chart for question 3: Pokemons with no multipliers
labels = ["With Multipliers", "Without Multipliers"]
sizes = [len(pokemon_list) - len(no_multipliers), len(no_multipliers)]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pokemons with and without Multipliers")
plt.axis('equal')
plt.show()
```

```python
#Q4 : Get all Pokemons that do not have more than 2 evolutions
evolution_filtered = [pokemon for pokemon in pokemon_list if len(pokemon.get('next_evolution', [])) <= 2]
for pokemon in evolution_filtered:
    print(pokemon['name'])
```

```python
# Bar plot for question 4: Pokemons with 2 or fewer evolutions
pokemon_names = [pokemon["name"] for pokemon in evolution_filtered]
evolution_counts = [len(pokemon.get("next_evolution", [])) for pokemon in evolution_filtered]

fig=plt.figure(figsize =(12,6))
plt.bar(pokemon_names, evolution_counts)
plt.xlabel("Pokemon")
plt.ylabel("Number of Evolutions")
plt.title("Pokemons with 2 or Fewer Evolutions")
plt.xticks(rotation=90, fontsize=0.8)
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()
plt.show()
```

```python
#Q5 : Get all Pokemons whose spawn time is less than 300 seconds
spawn_time_filtered = [pokemon for pokemon in pokemon_list if pokemon['spawn_time'] != 'N/A' and int(pokemon['spawn_time'].split(":")[0]) < 300]
for pokemon in spawn_time_filtered:
    print(pokemon['name'])
```

```python
# Bar plot for question 5: Pokemons with spawn time less than 300 seconds
pokemon_names = [pokemon["name"] for pokemon in spawn_time_filtered]
spawn_times = [int(pokemon["spawn_time"].split(":")[0]) for pokemon in spawn_time_filtered]

fig=plt.figure(figsize=(12,6))
plt.bar(pokemon_names, spawn_times)
plt.xlabel("Pokemon")
plt.ylabel("Spawn Time (Minutes)")
plt.title("Pokemons with Spawn Time Less than 300 Seconds")
plt.xticks(rotation=90,fontsize=0.8)
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()
plt.show()
```

```python
#Q6 : Get all Pokemon who have more than two types of capabilities
multi_type_filtered = [pokemon for pokemon in pokemon_list if len(pokemon['type']) > 1]
for pokemon in multi_type_filtered:
    print(pokemon['name'])
```

```python
# Bar plot for question 6: Pokemons with more than two types of capabilities
pokemon_names = [pokemon["name"] for pokemon in multi_type_filtered]
capability_counts = [len(pokemon["type"]) for pokemon in multi_type_filtered]

fig=plt.figure(figsize=(12,6))
plt.bar(pokemon_names, capability_counts)
plt.xlabel("Pokemon")
plt.ylabel("Number of Capabilities")
plt.title("Pokemons with More than Two Types of Capabilities")
plt.xticks(rotation=90,fontsize=0.8 )
plt.subplots_adjust(bottom=0.3)
plt.tight_layout()
plt.show()
```
