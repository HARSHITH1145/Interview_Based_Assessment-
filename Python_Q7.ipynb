```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("nasa_data.csv")
```

```python
data
```

# ● Get all the Earth meteorites that fell before the year 2000

```python
data['year']
```

```python
# i took the meteorites fell before year 2000 for the original data
data['year']=pd.to_datetime(data['year'])
meteorites = data[data['year'].dt.year < 2000]
meteorites
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data['name'],data['year'])
plt.xlabel('Names of meteorites')
plt.ylabel('Years')
plt.title('Earth meteorites that fell before the year 2000')
plt.grid(True)
plt.show()
```

# ● Get all the earth meteorites co-ordinates who fell before the year 1970

```python
# here i took the meteorites coordinates data which fell before the year 1970 from the original data 
coordinates = data[data['year'].dt.year < 1970][['name','reclong','reclat','coordinates']]
coordinates
```

```python
plt.figure(figsize=(10,10))
plt.scatter(coordinates['reclong'], coordinates['reclat'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title("Earth Meteorites Coordinates Before 1970")
plt.grid(True)
plt.show()
```

# ● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
data
```

```python
mass = data[data['mass'] > 10000]
mass
```

```python
plt.figure(figsize=(8,6))
plt.hist(mass['mass'], bins=10)
plt.xlabel('Mass(kg)*10^6')
plt.ylabel("Count of Meteorites")
plt.title('Earth Meteorites Mass > 10000kg')
plt.grid(True)
plt.show()
```
